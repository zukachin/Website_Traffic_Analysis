```
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load data from a CSV file
df = pd.read_csv('/content/daily-website-visitors.csv')

# Remove commas from numerical columns
numeric_columns = ['Page.Loads', 'Unique.Visits', 'First.Time.Visits', 'Returning.Visits']
for col in numeric_columns:
    df[col] = df[col].str.replace(',', '').astype(float)

# Select the columns you want to use for clustering
X = df[['Page.Loads', 'Unique.Visits', 'First.Time.Visits', 'Returning.Visits']]

# Choose the number of clusters (K)
k = 3

# Initialize the K-Means model
kmeans = KMeans(n_clusters=k, random_state=0)

# Fit the model to your data
kmeans.fit(X)

# Get cluster labels for each data point
cluster_labels = kmeans.labels_

# Add the cluster labels to your DataFrame
df['Cluster'] = cluster_labels

# Visualize the clusters (for example, 2D scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(df['Page.Loads'], df['Unique.Visits'], c=cluster_labels, cmap='rainbow')
plt.xlabel('Page Loads')
plt.ylabel('Unique Visits')
plt.title('K-Means Clustering')
plt.show()
```

```
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import metrics

# Load your data from a CSV file and preprocess it as mentioned before
df = pd.read_csv('/content/daily-website-visitors.csv')
numeric_columns = ['Page.Loads', 'Unique.Visits', 'First.Time.Visits', 'Returning.Visits']
for col in numeric_columns:
    df[col] = df[col].str.replace(',', '').astype(float)

# Select the columns for clustering
X = df[['Page.Loads', 'Unique.Visits', 'First.Time.Visits']]

# Choose the number of clusters (K)
k = 3

# Initialize the K-Means model
kmeans = KMeans(n_clusters=k, random_state=0)

# Fit the model to your data and get cluster labels
cluster_labels = kmeans.fit_predict(X)

# Calculate Inertia
inertia = kmeans.inertia_

# Calculate Silhouette Score
silhouette_score = metrics.silhouette_score(X, cluster_labels)

# Calculate Davies-Bouldin Index
davies_bouldin = metrics.davies_bouldin_score(X, cluster_labels)

# Calculate Calinski-Harabasz Index
calinski_harabasz = metrics.calinski_harabasz_score(X, cluster_labels)

# Calculate Adjusted Rand Index
true_labels = df['Page.Loads']  # Replace 'True_Labels' with your actual labels if available
adjusted_rand_index = metrics.adjusted_rand_score(true_labels, cluster_labels)

# Calculate Normalized Mutual Information
normalized_mutual_info = metrics.normalized_mutual_info_score(true_labels, cluster_labels)

# Calculate Fowlkes-Mallows Index
fowlkes_mallows = metrics.fowlkes_mallows_score(true_labels, cluster_labels)

# Print the evaluation metrics
print("Inertia: {:.2f}".format(inertia))
print("Silhouette Score: {:.2f}".format(silhouette_score))
print("Davies-Bouldin Index: {:.2f}".format(davies_bouldin))
print("Calinski-Harabasz Index: {:.2f}".format(calinski_harabasz))
print("Adjusted Rand Index: {:.2f}".format(adjusted_rand_index))
print("Normalized Mutual Information: {:.2f}".format(normalized_mutual_info))
print("Fowlkes-Mallows Index: {:.2f}".format(fowlkes_mallows))
```
